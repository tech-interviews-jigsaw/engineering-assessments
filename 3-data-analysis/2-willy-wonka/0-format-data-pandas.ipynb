# Loading Data

```python
import pandas as pd
df = pd.read_csv('solutions_simulation_willy_wonka.csv')
df.columns = [col.lower() for col in df.columns]
```

```python
# list_sku	list_sku_category	list_sku_flavor

first_sku = df['list_sku'].str.split(',').str[0]
second_sku = df['list_sku'].str.split(',').str[1]

first_sku_cat = df['list_sku_category'].str.split(',').str[0]
second_sku_cat = df['list_sku_category'].str.split(',').str[1]

first_sku_flavor = df['list_sku_flavor'].str.split(',').str[0]
second_sku_flavor = df['list_sku_flavor'].str.split(',').str[1]

df_dropped = df.drop(columns = ['list_sku', 'list_sku_category', 'list_sku_flavor'])

updated_df = df_dropped.assign(first_sku = first_sku, second_sku = second_sku,
 first_sku_cat = first_sku_cat, second_sku_cat = second_sku_cat,
  first_sku_flavor = first_sku_flavor, second_sku_flavor = second_sku_flavor)
```

```python
updated_df_first = df_dropped.assign(list_sku = first_sku, 
 list_sku_category = first_sku_cat,
  list_sku_flavor = first_sku_flavor)

updated_df_first[:2]
```

```python
second_sku_df = updated_df[~updated_df['second_sku'].isna()]
updated_second_sku_df = second_sku_df.drop(columns = ['first_sku', 'first_sku_cat', 'first_sku_flavor'])

updated_second_sku_df_renamed = updated_second_sku_df.rename(columns={"second_sku": "list_sku", 
"second_sku_cat": "list_sku_category", "second_sku_flavor": "list_sku_flavor"})

updated_second_sku_df_renamed[:2]
```

```python
combined_df = pd.concat([updated_second_sku_df_renamed, updated_df_first], ignore_index=True)

combined_df[:1]
```
